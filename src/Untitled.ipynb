#### Get all "calendar.csv" ( feb, mar, apr)

```python
my_usecols=['id', 
'zipcode',
'transit',
'bedrooms',
'beds',
'review_scores_rating',
'number_of_reviews',
'neighbourhood',
'name',
'latitude',
'longitude',
'last_review',
'instant_bookable',
'host_since',
'host_response_rate',
'host_identity_verified',
'host_has_profile_pic',
'first_review',
'description',
'city',
'cancellation_policy',
'bathrooms',
'accommodates',
'amenities',
'room_type',
'property_type',
'price',
'availability_365',
'minimum_nights']
location1 = r'../../data/february/listings.csv'
feb_listing = pd.read_csv(location1, usecols=my_usecols)
location2 = r'../../data/march/listings.csv'
mar_listing = pd.read_csv(location2,usecols=my_usecols)
location3 = r'../../data/april/listings.csv'
apr_listing = pd.read_csv(location3,usecols=my_usecols)
location4  = r'../../data/february/listings0.csv'
feb_listing0 = pd.read_csv(location4)
location5  = r'../../data/march/listings0.csv'
mar_listing0 = pd.read_csv(location5)
location6  = r'../../data/april/listings0.csv'
apr_listing0 = pd.read_csv(location6)

feb_listing0_set = set(feb_listing0.columns)
my_usecols_set = set(my_usecols)

not_in_list = feb_listing0_set - feb_listing0_set.intersection(my_usecols_set)


```

```python
all_listings = pd.concat([feb_listing,mar_listing, apr_listing])

all_listings0 = pd.concat([feb_listing0,mar_listing0, apr_listing0])

all_listings0 = all_listings0.drop(columns=not_in_list, axis = 1)

all_listings.append(all_listings0)

all_listings = all_listings.drop_duplicates()
all_listings

```
