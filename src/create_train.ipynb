```python
import pandas as pd #this is how I usually import pandas
import sys #only needed to determine Python version number
```

```python
my_usecols= ['id', 
'zipcode',
'transit',
'bedrooms',
'beds',
'review_scores_rating',
'number_of_reviews',
'neighbourhood',
'name',
'latitude',
'longitude',
'last_review',
'instant_bookable',
'host_since',
'host_response_rate',
'host_identity_verified',
'host_has_profile_pic',
'first_review',
'description',
'city',
'cancellation_policy',
'bathrooms',
'accommodates',
'amenities',
'room_type',
'property_type',
'price',
'availability_365',
'minimum_nights']

location1 = r'../../data/february/listings.csv'
feb_listing = pd.read_csv(location1, usecols=my_usecols)
feb_review = pd.read_csv('../../data/february/reviews.csv', usecols=['listing_id', 'date', 'comments'])

location2 = r'../../data/march/listings.csv'
mar_listing = pd.read_csv(location2,usecols=my_usecols)
mar_review = pd.read_csv('../../data/march/reviews.csv', usecols=['listing_id', 'date', 'comments'])

location3 = r'../../data/april/listings.csv'
apr_listing = pd.read_csv(location3,usecols=my_usecols)
apr_review = pd.read_csv('../../data/april/reviews.csv', usecols=['listing_id', 'date', 'comments'])

my_usecols.append('comments')
my_usecols.append('month')

feb_listing.insert(loc = len(feb_listing.columns), column='month', value='02_February')
feb_listing = feb_listing.merge(feb_review, left_on=['id', 'last_review'], right_on=['listing_id', 'date']).filter(items=my_usecols)

mar_listing.insert(loc = len(mar_listing.columns), column='month', value='03_March')
mar_listing = mar_listing.merge(mar_review, left_on=['id', 'last_review'], right_on=['listing_id', 'date']).filter(items=my_usecols)

apr_listing.insert(loc = len(apr_listing.columns), column='month', value='04_April')
apr_listing = apr_listing.merge(apr_review, left_on=['id', 'last_review'], right_on=['listing_id', 'date']).filter(items=my_usecols)

```

```python
all_listings = pd.concat([feb_listing,mar_listing, apr_listing])
```

```python
all_listings = all_listings.replace(to_replace = {'price':'\$'}, value='', regex=True)
all_listings = all_listings.replace(to_replace = {'price':'\,'}, value='', regex=True)
```

```python
all_listings = all_listings.astype({'price':'float64'})
```
